Calcular a área do retângulo

```python
# Entrada
comprimento = float(input("Digite o comprimento do retângulo:"))
altura = float(input("Digite a altura do retângulo:"))

# Processamento

area = comprimento * altura


# Saída

print (f"O total da área do retângulo é:{area}")
```
